```python
# 70. Climbing Stairs
# Solved
# Easy
# Topics
# Companies
# Hint
# You are climbing a staircase. It takes n steps to reach the top.

# Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

 

# Example 1:

# Input: n = 2
# Output: 2
# Explanation: There are two ways to climb to the top.
# 1. 1 step + 1 step
# 2. 2 steps
# Example 2:

# Input: n = 3
# Output: 3
# Explanation: There are three ways to climb to the top.
# 1. 1 step + 1 step + 1 step
# 2. 1 step + 2 steps
# 3. 2 steps + 1 step


def climbStairs(n):
    if n <= 2:         # Base case: n=1 has 1 way, n=2 has 2 ways
        return n
    else:              # Recursive case: sum of the ways to get to n-1 and n-2
        return climbStairs(n-1) + climbStairs(n-2)


n = 6
x =climbStairs(n)
print(x)



# Detailed Explanation:

# Base Case (n <= 2):
# If n is 1, there is 1 way to climb the stairs (just take 1 step).
# If n is 2, there are 2 ways: (1+1) or (2).

# Recursive Case:
# For n > 2, the number of distinct ways to climb n steps is the sum of the ways to 
# climb n-1 steps and n-2 steps. This is because from step n-1, you can take a single 
# step to reach n, and from step n-2, you can take two steps to reach n.
# Example Walkthrough:
# Let's say n = 4. Here's how the recursion unfolds:

# climbStairs(4) = climbStairs(3) + climbStairs(2)
# climbStairs(3) = climbStairs(2) + climbStairs(1)
# climbStairs(2) = 2 (base case)
# climbStairs(1) = 1 (base case)
```

```python
# 172. Factorial Trailing Zeroes
# Medium
# Topics
# Companies
# Given an integer n, return the number of trailing zeroes in n!.

# Note that n! = n * (n - 1) * (n - 2) * ... * 3 * 2 * 1.

 

# Example 1:

# Input: n = 3
# Output: 0
# Explanation: 3! = 6, no trailing zero.
# Example 2:

# Input: n = 5
# Output: 1
# Explanation: 5! = 120, one trailing zero.
# Example 3:

# Input: n = 0
# Output: 0



def factorial(n):
	if n == 0 or n == 1:
		return 1
	else:
		return n*factorial(n-1)

n = 3
x = factorial(n)
print(x)


'''
Detailed Explanation:
Base Case (n == 0 or n == 1):

Factorial of 0 (0!) and 1 (1!) is always 1. This is the stopping condition for the recursion.
Recursive Case:

If n > 1, the function calls itself with n-1 until it reaches 1. 
Each recursive call returns the factorial of n-1, which is then multiplied by n itself.
Example Walkthrough:
Let's say n = 4. Here's how the recursion unfolds:

factorial(4) = 4 * factorial(3)
factorial(3) = 3 * factorial(2)
factorial(2) = 2 * factorial(1)
factorial(1) = 1 (base case)
'''

########################### if we put n > 1000 then we will hit recursion error ########################### 

# The error you're encountering is a RecursionError which happens because the Python interpreter has a maximum 
# recursion depth limit to prevent a stack overflow. 
# By default, this limit is set to 1000 in Python. 
# When you try to calculate the factorial of 3000 using a recursive approach, 
# the function calls itself too many times, exceeding this limit.


#Another approach without recursion

def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

n = 3
x = factorial(n)
print(x)
```

```python
# 509. Fibonacci Number
# The Fibonacci numbers, commonly denoted F(n) form a sequence, called the Fibonacci sequence, such that each number is the sum of the two preceding ones, starting from 0 and 1. That is,

# F(0) = 0, F(1) = 1
# F(n) = F(n - 1) + F(n - 2), for n > 1.
# Given n, calculate F(n).

 

# Example 1:

# Input: n = 2
# Output: 1
# Explanation: F(2) = F(1) + F(0) = 1 + 0 = 1.
# Example 2:

# Input: n = 3
# Output: 2
# Explanation: F(3) = F(2) + F(1) = 1 + 1 = 2.
# Example 3:

# Input: n = 4
# Output: 3
# Explanation: F(4) = F(3) + F(2) = 2 + 1 = 3.





def fib(n):
	if n <= 1:
		return n
	else:
		return fib(n-1) + fib(n-2)
n = 4
x = fib(n)
print(x)





# Detailed Explanation:

# Base Case (n <= 1):
# If n is 0 or 1, we return n. This is because we know the first two Fibonacci numbers: F(0) = 0 and F(1) = 1.

# Recursive Case:
# For any n > 1, we recursively calculate the Fibonacci numbers for n-1 and n-2, and then add them together. 
# This is based on the recurrence relation: F(n) = F(n-1) + F(n-2)


# Input: n = 4
# Output: 3
# Explanation: 
# F(4) = F(3) + F(2)
#      = (F(2) + F(1)) + (F(1) + F(0))
#      = ((F(1) + F(0)) + 1) + (1 + 0)
#      = (1 + 0 + 1) + (1 + 0)
#      = 2 + 1
#      = 3
```
